```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://finance.naver.com/api/sise/etfItemList.nhn?etfType=0&targetColumn=market_sum&sortOrder=desc'
url
```

```python
#pd.read_html(url)
```

```python
response = requests.get(url)
response
```

```python
response.text
```

```python
#response.json()
```

```python
etf_json = response.json()
etf_json
```

```python
etfItemList = etf_json['result']['etfItemList']
etfItemList
```

```python
len(etfItemList)
```

```python
etf_df = pd.DataFrame(etfItemList)
etf_df.shape
```

```python
etf_df.head(10)
```

### 파일 만들기

```python
import datetime

today = datetime.datetime.today()
today = today.strftime('%y-%m-%d')
today
```

```python
file_name = f'etf_{today}_raw.csv'
file_name
```

```python
etf_df['itemcode'] = etf_df['itemcode'].astype(object)

etf_df.to_csv(file_name, index = False)
```

```python
etf_df = pd.read_csv(file_name, dtype = {'itemcode': object})
etf_df.head()
```
