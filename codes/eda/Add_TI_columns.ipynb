```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

etf_name = 'TIGER_반도체'
df = pd.read_csv(f'{etf_name}.csv')
df.head()
```

```python
# '날짜' 열의 데이터 타입을 datetime으로 변환
df['날짜'] = pd.to_datetime(df['날짜'])

# '날짜' 열을 기준으로 오름차순으로 정렬
df = df.sort_values(by='날짜')
```

```python
df = pd.DataFrame(df[['날짜', '종가']])
```

```python
df['diff_5'] = df['종가'].pct_change(5)
df['diff_20'] = df['종가'].pct_change(20)

df['ma_5'] = df['종가'].rolling(window=5).mean()
df['ma_20'] = df['종가'].rolling(window=20).mean()
```

```python
df.head()
```

```python
df = df.iloc[20:]
df.head()
```

```python
# 이건 ETF 종목명에 따라 변경하면 됨
df.to_csv(f'{etf_name}_TI.csv')
```
