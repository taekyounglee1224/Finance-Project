### 데이터프레임 분리

결측치 개수에 따라 두 개의 데이터프레임으로 분리

결측치가 많은 데이터 :  NAN_temp_df.csv

결측치가 적은 데이터 :  temp_df.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp_df = pd.read_csv('/temp_data.csv', index_col = 0)
temp_df.head()
```

```python
#NaN값이 많은 열만 추출
NaN_temp_df = temp_df[['stateDt', 'clearCnt', 'examCnt', 'careCnt', 'resutlNegCnt', 'accExamCnt', 'accExamCompCnt', 'accDefRate']]
```

```python
#stateDt를 정수형으로 변환
NaN_temp_df['stateDt'] = NaN_temp_df['stateDt'].astype('int')
```

```python
#stateDt을 기준으로 오름차순으로 정렬
NaN_temp_df = NaN_temp_df.sort_values(by='stateDt')
```

```python
NaN_temp_df
```

```python
#데이터프레임을 csv로 변환 후 저장
NaN_temp_df.to_csv('/NaN_temp_df.csv')
```

```python
temp_df = pd.read_csv('/temp_data.csv', index_col = 0)
temp_df.head()
```

```python
#NaN값이 없는 열만 추출
temp_df = temp_df[['stateDt', 'decideCnt', 'deathCnt']]
```

```python
#stateDt을 기준으로 오름차순으로 정렬
temp_df = temp_df.sort_values(by='stateDt')
```
